```python
texts_paths = list((pathlib.Path(".") / "corpus").glob("*"))
print(texts_paths)
```

```python
convert = {
    "cx": "ĉ",
    "gx": "ĝ",
    "hx": "ĥ",
    "jx": "ĵ",
    "sx": "ŝ",
    "ux": "ŭ"
}
alphabet = "abcĉdefgĝhĥijĵklmnoprsŝtuŭvz ."
def preprocess_text(text: str):
    text = text.lower()
    for k, v in convert.items():
        text = text.replace(k, v)
    text = "".join(c for c in text if c in alphabet)
    text = re.sub(r"\s+", " ", text)
    text = re.sub(r"[\s\.]*\.[\s\.]*", ". ", text, flags=re.MULTILINE)
    return text
    
```

```python
texts = ""
for text_path in texts_paths:
    print(text_path)
    with open(text_path) as f:
         text = preprocess_text(f.read())
    with open(text_path.parent.parent / "processed" / text_path.stem, "w") as f:
        f.write(text)
    texts += text
```

```python
# Build the model.
text_model = markovify.NewlineText("\n".join(texts.split(".")))
```
